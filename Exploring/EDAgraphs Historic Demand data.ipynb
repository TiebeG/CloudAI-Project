#### Load data

```python
import pandas as pd
demand_df = pd.read_csv("..\Data\demand_prices\cleaned_data_neso_demand_2001_2025.csv", compression='gzip')
display(demand_df)
demand_df.info()
```

```python
# Create a copy of the original dataframe
numeric_df = demand_df.copy()

# Fill all NaN values with 0
numeric_df = numeric_df.fillna(0)
display(numeric_df)
numeric_df.info()
```

### Graphs

Graph 1: National demand over time

```python
import matplotlib.pyplot as plt
import pandas as pd

# Convert date to datetime if not already
demand_df['settlement_date'] = pd.to_datetime(demand_df['settlement_date'])

# Sort just in case
demand_df = demand_df.sort_values('settlement_date')

plt.figure(figsize=(15,5))
plt.plot(demand_df['settlement_date'], demand_df['nd'])
plt.title("National Demand (ND) over time")
plt.xlabel("Date")
plt.ylabel("ND (MW)")
plt.show()
```

Graph 2: Average ND per settlement period (daily demand curve)

```python
daily_curve = demand_df.groupby('settlement_period')['nd'].mean()

plt.figure(figsize=(12,5))
plt.plot(daily_curve.index, daily_curve.values)
plt.title("Average ND by Settlement Period")
plt.xlabel("Settlement period (1–48)")
plt.ylabel("Average ND (MW)")
plt.show()
```

Graph 3: Embedded Solar & Wind generation trends

```python
solar = demand_df.groupby(demand_df['settlement_date'].dt.year)['embedded_solar_generation'].mean()
wind  = demand_df.groupby(demand_df['settlement_date'].dt.year)['embedded_wind_generation'].mean()

plt.figure(figsize=(12,5))
plt.plot(solar.index, solar.values, label='Solar')
plt.plot(wind.index, wind.values, label='Wind')
plt.title("Avg Solar and Wind Generation per Year")
plt.xlabel("Year")
plt.ylabel("Generation (MW)")
plt.legend()
plt.show()
```

Graph 4: Distribution of flows (example: NEMO flow)

```python
plt.figure(figsize=(8,4))
plt.hist(demand_df['nemo_flow'], bins=100)
plt.title("Distribution of NEMO Interconnector Flow")
plt.xlabel("Flow (MW)")
plt.ylabel("Count")
plt.show()
```

```python
flow_cols = ['ifa_flow', 'ifa2_flow', 'nemo_flow', 'britned_flow', 'east_west_flow']
demand_df[flow_cols].plot(kind='box', figsize=(12,6))
plt.title("Interconnector Flow Distributions")
plt.show()
```

### 2. Correlation Analysis

Compute correlation matrix:

```python
corr = numeric_df.corr()
corr
```

```python
plt.figure(figsize=(14,10))
plt.imshow(corr, cmap='coolwarm', interpolation='nearest')
plt.colorbar(label='Correlation')

plt.xticks(range(len(corr.columns)), corr.columns, rotation=90)
plt.yticks(range(len(corr.columns)), corr.columns)

plt.title("Correlation Heatmap")
plt.tight_layout()
plt.show()
```
